## Импорт данных<a class="anchor" id="data-import"></a>

```python
import pandas as pd
import os
from sqlalchemy import create_engine
```

```python
db_config = {'user': '---', # имя пользователя
            'pwd': '---', # пароль
            'host': '---',
            'port': ---, # порт подключения
            'db': '---'} # название базы данных

connection_string = 'postgresql://{}:{}@{}:{}/{}'.format(db_config['user'],
                                                db_config['pwd'],
                                                db_config['host'],
                                                db_config['port'],
                                                db_config['db'])

engine = create_engine(connection_string) 
```

```python
query = '''
            SELECT * FROM dash_visits
        '''
```

```python
visits = pd.io.sql.read_sql(query, con = engine)
visits
```

```python
visits.to_csv('dash_visits.csv')
```

#### Как много карточек генерируют источники с разными темами?
- Больше всего карточек генерируют источники по "общечеловеческим темам”: отношения, советы, путешествия;
- Абсолютный лидер во взаимодействию: источник путешествия и карточка рассказы. Трэвел-блоги в топе.

#### Как соотносятся темы карточек и темы источников?
- Наиболее популярные источники и карточки тематически взаимосвязаны. Например: политика и общество популярное сочетание, а женская психология и строительство — нет. Возможно это связано с тем как расставляются теги;
- Есть явные лидеры среди источников которые генерируют карточки почти во всем темам, например семейные отношения, путешествия, полезные советы;
